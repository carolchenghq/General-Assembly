```python
import seaborn as sns
import matplotlib.pyplot as plt

# plot the heatmap
def correlation_heat_map(df):
    corrs = df.corr()

    # Set the default matplotlib figure size:
    fig, ax = plt.subplots(figsize=(14,9))

    # Generate a mask for the upper triangle (taken from seaborn example gallery)
    mask = np.zeros_like(corrs, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True

    # Plot the heatmap with seaborn.
    # Assign the matplotlib axis the function returns. This will let us resize the labels.
    ax = sns.heatmap(corrs, mask=mask, annot=True, cmap="coolwarm")

    # Resize the labels.
    ax.set_xticklabels(ax.xaxis.get_ticklabels(), fontsize=12, rotation=80)
    ax.set_yticklabels(ax.yaxis.get_ticklabels(), fontsize=12, rotation=0)

    # If you put plt.show() at the bottom, it prevents those useless printouts from matplotlib.
    plt.show()

correlation_heat_map(df_bank)
```

```python
# From the correlation matrix and the heatmap above, it looks the target y has strong positive correlation with 
# 'last_contact_duration' and 'previous campaign_contact', and has strong negative correlation with 
# 'days_from_last_contact', 'employment_variation_rate', 'euribor_3m_rate' and 'num_of_employees'
```

### 1. 'age'
- Question1: What is the ages distribution for people who has purchased the product ?
- Question2: How people's buy will changed along ages ?

```python
# Plot the distribution of 'age'

fig, axes = plt.subplots(1,3, figsize=(14,4), squeeze=False)

df_bank['age'].plot(kind='hist', color='skyblue', ax=axes[0][0])
df_bank[df_bank['y']==1]['age'].plot(kind='hist', color='lightblue', ax=axes[0][1])
df_bank[df_bank['y']==0]['age'].plot(kind='hist',color='steelblue', ax=axes[0][2])

axes[0][0].set_title('Age distribution of all clients', fontsize=13)
axes[0][1].set_title('Age distribution of purchased clients', fontsize=13)
axes[0][2].set_title('Age distribution of unpurchased clients', fontsize=13)
```

```python
# Add a new column 'age_group' for analysis purpose
def ageGroup(age):
    if age < 20:
        return '0-20'
    elif age <= 30:
        return '20-30'
    elif age <= 40:
        return '31-40'
    elif age <= 50:
        return '41-50'
    elif age <= 60:
        return '51-60'
    elif age <= 70:
        return '61-70'
    else:
        return 'over 70'
df_age_grouped = df_bank.copy()
df_age_grouped['age_group'] = df_age_grouped.apply(lambda row: ageGroup(row.age), axis=1)
df_age_grouped.head()
```

```python
fig, axes = plt.subplots(1,3, figsize=(15,4), squeeze=False)

df_age_grouped['age_group'].value_counts().sort_index().plot(kind='bar', rot=0, ax=axes[0][0], color='indianred')
df_age_grouped[df_age_grouped['y']==1]['age_group'].value_counts().sort_index().plot(kind='bar', rot=0,ax=axes[0][1], color='seagreen')
df_age_grouped[df_age_grouped['y']==0]['age_group'].value_counts().sort_index().plot(kind='bar', rot=0,ax=axes[0][2], color='steelblue')

axes[0][0].set_title('Age groups of all clients', fontsize=13)
axes[0][1].set_title('Age groups of purchased clients', fontsize=13)
axes[0][2].set_title('Age groups of unpurchased clients', fontsize=13)

axes[0][0].xaxis.set_tick_params(labelsize=11)
axes[0][1].xaxis.set_tick_params(labelsize=11)
axes[0][2].xaxis.set_tick_params(labelsize=11)



#x = ['0-20','20-30','31-40','41-50','51-60','61-70','over 70']
#y = [71,7124,15956,9963,6097,475,412]

#for i, v in enumerate(y):
#    ax.text(v + 3, i + .25, str(v), color='blue', fontweight='bold')
```

```python
ax_pie = df_age_grouped[df_age_grouped['y']==1]['age_group'].value_counts().plot(kind='pie', cmap='Reds',
                                                                                 explode=[0.05,0.05,0.05,0.05,0.05,0.05,0.05], 
                                                                                 autopct='%1.0f%%')
ax_pie.set_ylabel('')
ax_pie.set_title('Ages distribution of the clients \nwho has purchased the product', fontsize=14)
```

```python
# Plot how clients number change across all ages.
# Have a further inspection in the age group '20-30' and '30-40', beause clients number obviously changed in these 
# two age groups.
fig, axes = plt.subplots(3,1, figsize=(10,14), squeeze=False)

df_20to30 = df_age_grouped[df_age_grouped['y']==1][df_age_grouped['age_group']=='20-30']
df_30to40 = df_age_grouped[df_age_grouped['y']==1][df_age_grouped['age_group']=='31-40']


df_bank[df_bank['y']==1]['age'].value_counts().sort_index().plot(ax=axes[0][0], color='teal')
df_20to30['age'].value_counts().sort_index().plot(ax=axes[1][0], color='lightseagreen')
df_30to40['age'].value_counts().sort_index().plot(ax=axes[2][0], color='lightseagreen')

axes[0][0].set_title('The change of purchased clients numbers across all ages', fontsize=16)
axes[1][0].set_title('The change of purchased clients numbers across 20-30', fontsize=16)
axes[2][0].set_title('The change of purchased clients numbers across 30-40', fontsize=16)
```

```python
# - From the above charts, the clients who has purchased the product concentrated on age from 20 to 40.
# - The No.1 clients age group is 31-40. Around one third of the purchases made by the clients in this age group.
# - The 2nd largest clients age group is 20-30, which made 24% of the purchases. 
# - Number of purchase clients increased quickly from age 23, reached the peek at age 31, and then after that droped 
#   down quickly until age 40. The purchases stayed relatively stable among 40-60, and remarkably decreased from 
#   around 60.
```

```python
# Reason why people in 20-30 more tend to subscribe fixed deposit than people in other age groups?
# - Started to get a job and earn money after around 23
# - Didn't have large sum of expenditures yet
# - Tended to deposit money for future use
# - Didn't pay much attention on investment

# Reason why people gets less and less interested in fixed deposit after age 31 ?
# - Many people started having a family after around 30, their expenditures greatly increased. (expenditure on wedding,
#   homeloan, expense on raising a child, etc.)
# - As people age grew, they had more ideas of investment.(Other than simply put their money to fixed term deposit,
#   they now knew more about options of investing their money.)
```

```python
# As perfectly or highly correlation exist among these variables, this can cause “Multicollinearity” which will impact
# the performance of modeling. 

# Decision trees and boosted trees algorithms are immune to multicollinearity by nature. When they decide to split, 
# the tree will choose only one of the perfectly correlated features. 

# However, other algorithms like Logistic Regression or Linear Regression are not immune to that problem. Then I will 
# need to fix it before training the model.
```

```python
# Strategies to deal with multicollinearity:
# - Eliminate one of the perfectly correlated features
# - Use a dimension reduction algorithm such as Principle Component Analysis (PCA)
# - Use algorithms that are immune to nulticollinearity
```

```python
# Study points: Spearman and Pearson. 
# - Pearson Correlation Coefficient can be used with continuous variables that have a linear relationship.
# - Spearman Correlation Coefficient can be used if the variables have a non-linear relationship. It can also be 
#   used with ordinal categorical variables. (scipy.stats.spearmanr(X,Y))
```

## Visualization for categorical variables

```python
# Is there a good way to visualize categorical variables? Let's try Parallel Coordinates from Yellowbrick
```

```python
# Firstly create a subset that only contains expected categorical variables

df_categorical = df_bank[['job', 'marital', 'education', 'credit_default', 'housing_loan', 
                'personal_loan', 'contact', 'previous_campaign_outcome', 'y']].copy()

# Turn the numerical columns back to object type so that I can use Parallel Coordinates
df_categorical['housing_loan'] = df_categorical['housing_loan'].map(lambda x: 'yes' if x ==1 else 'no')
df_categorical['personal_loan'] = df_categorical['personal_loan'].map(lambda x: 'yes' if x ==1 else 'no')
df_categorical['y'] = df_categorical['y'].map(lambda x: 'yes' if x ==1 else 'no')
```

```python
from yellowbrick.features import ParallelCoordinates

# Define X and y
X = df_categorical[['job', 'marital', 'education', 'credit_default', 'housing_loan', 
                'personal_loan', 'contact', 'previous_campaign_outcome']]
y = df_categorical['y']

# Specify the features of interest and the classes of the target
features = ['job', 'marital', 'education', 'credit_default', 'housing_loan', 
             'personal_loan', 'contact', 'previous_campaign_outcome']
classes = ['yes', 'no']

# Instantiate the visualizer
visualizer = ParallelCoordinates(classes=classes, features=features, sample=0.1, shuffle=True)

# Fit and transform the data to the visualizer
visualizer.fit_transform(X, y)

# Finalize the title and axes then display the visualization
visualizer.show()

# As the data is imbalance, the graph doesn't look good.
```
